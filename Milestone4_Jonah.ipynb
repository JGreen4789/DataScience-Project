```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
SAT_Scores = pd.read_csv("New_SAT__College_Board__2010_School_Level_Results.csv")
```

```python
SAT_Scores.head()
```

```python
SAT_Scores=SAT_Scores.dropna()
```

```python
SAT_Scores
```

```python
SAT_Scores.plot.scatter(x="Number of Test Takers",y="Critical Reading Mean")
plt.title("Critical Reading Mean by Number of Test Takers ")
plt.xlabel("Number of Test Takers")
plt.ylabel("Critical Reading Mean")
```

```python
SAT_Scores.plot.scatter(x="Number of Test Takers",y="Mathematics Mean")
plt.title("Mathematics Mean by Number of Test Takers")
plt.xlabel("Number of Test Takers")
plt.ylabel("Mathematics Mean")
```

```python
SAT_Scores.plot.scatter(x="Number of Test Takers",y="Writing Mean")
plt.title("Writing Mean by Number of Test Takers")
plt.xlabel("Number of Test Takers")
plt.ylabel("Writing Mean")
```
